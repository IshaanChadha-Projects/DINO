## Left shift

```python
import os
##from keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
from matplotlib import pyplot
from PIL import UnidentifiedImageError

# Define the folder containing the images
folder_path = 'archive-2/trex/'

# Get a list of all image filenames in the folder
image_files = os.listdir(folder_path)

# Create a directory to store augmented images
if not os.path.exists('left_shift_trex_images'):
    os.makedirs('left_shift_trex_images')

# create image data augmentation generator
datagen = ImageDataGenerator(height_shift_range=[-0.2, -0.2])

# Loop through each image file
for filename in image_files:
    # Load the image
    try:
        img = load_img(os.path.join(folder_path, filename))
    
    except UnidentifiedImageError:
        # Skip the problematic file and continue to the next one
        print(f"Skipping file {filename} due to UnidentifiedImageError")
        continue
    # Convert the image to an array
    data = img_to_array(img)
    # Expand the dimensions to create a batch of size 1
    samples = data.reshape((1,) + data.shape)
    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    
    # Generate and save augmented images
    for i in range(1):
        # generate batch of images
        batch = next(it)
        # convert to unsigned integers for viewing
        image = batch[0].astype('uint8')
        # Save the augmented image
        augmented_filename = 'left_shift_' + filename.split('.')[0] + '_' + str(i) + '.jpg'
        pyplot.imsave(os.path.join('left_shift_trex_images', augmented_filename), image)
```
